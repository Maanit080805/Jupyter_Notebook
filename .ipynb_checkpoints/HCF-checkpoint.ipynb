```python
class Hcf_Lcm:
    
    @classmethod
    def start(cls):
        print("HCF and LCM Caculator")
        cls.frc = input("Do you have fractions? (y/n): ")
    
    
```

```python
    @classmethod
    def ques_int(cls):
        cls.yn = int(input("How many numbers do you have? (2/3): "))
    
    
```

```python
    def ques_frc(cls):
        cls.inp = int(input("How many fractions do you have? (2/3): "))
    
```

```python
    @classmethod
    def entry(cls):
        cls.enter1 = int(input("Enter your 1st number: "))
        cls.enter2 = int(input("Enter your 2nd number: "))
    
    @classmethod
    def entry_2(cls):
        cls.enter1 = int(input("Enter your 1st number: "))
        cls.enter2 = int(input("Enter your 2nd number: "))
        cls.enter3 = int(input("Enter your 3rd number: "))
    
    @classmethod
    def entry_frc_1(cls):
        cls.num1 = int(input("Enter numerator of 1st fraction: "))
        cls.den1 = int(input("Enter denominator of 1st fraction: "))
        cls.num2 = int(input("Enter numerator of 2nd fraction: "))
        cls.den2 = int(input("Enter denominator of 2nd fraction: "))
    
    @classmethod
    def entry_frc_2(cls):
        cls.num1 = int(input("Enter numerator of 1st fraction: "))
        cls.den1 = int(input("Enter denominator of 1nd fraction: "))
        cls.num2 = int(input("Enter numerator of 2nd fraction: "))
        cls.den2 = int(input("Enter denominator of 2nd fraction: "))
        cls.num3 = int(input("Enter numerator of 3rd fraction: "))
        cls.den3 = int(input("Enter denominator of 3rd fraction: "))


        
    
    
```

```python
    @classmethod
    def hcf_2(cls):
        
        count = []
        count_2 = []
        result = []
        if cls.enter1 == cls.enter2:
            print("HCF is>", cls.enter2)
        else:
            for i in range(1,cls.enter1+1):
                if cls.enter1 % i == 0:
                    count.append(i)
                else:
                    pass
            for i in range(1,cls.enter2+1):
                if cls.enter2 % i == 0:
                    count_2.append(i)
                else:
                    pass
            for x in count:
                for y in count_2:
                    if y == x:
                        result.append(y)
                    else:
                        pass
            cls.hcf_ = max(result)
            print("HCF is>",cls.hcf_)
    
   
```

```python
    @classmethod
    def lcm_2(cls):
        if cls.enter1 == cls.enter2:
            print("LCM is>", cls.enter2)
        else:
            lcm_ = (cls.enter1*cls.enter2)//cls.hcf_
            print("LCM is>",lcm_)
    
   
```

```python
    @classmethod
    def hcf_3(cls):
        how = []
        how_2 = []
        how_3 = []
        ab = []
        bc = []
        ca = []
        result_2 = []
        if cls.enter1 == cls.enter2 == cls.enter3:
            print("HCF is>", cls.enter2)
        else:
            for i in range(1,cls.enter1+1):
                if cls.enter1 % i == 0:
                    how.append(i)
                else:
                    pass
            for i in range(1,cls.enter2+1):
                if cls.enter2 % i == 0:
                    how_2.append(i)
                else:
                    pass
            for i in range(1,cls.enter3+1):
                if cls.enter3 % i == 0:
                    how_3.append(i)
                else:
                    pass
            for x in how:
                for y in how_2:
                    if x == y:
                        ab.append(y)
                    else:
                        pass
            for x in how_2:
                for y in how_3:
                    if x == y:
                        bc.append(y)
                    else:
                        pass
            for x in how:
                for y in how_3:
                    if x == y:
                        ca.append(y)
                    else:
                        pass
            for x in how:
                for y in how_2:
                    for z in how_3:
                        if y == x == z:
                            result_2.append(y)
                        else:
                            pass
            cls.hcf_ab = max(ab)
            cls.hcf_bc = max(bc)
            cls.hcf_ca = max(ca)
            cls.answer = max(result_2)
            print("HCF is>",cls.answer)
    
    
```

```python
    @classmethod
    def lcm_3(cls):
        item = cls.enter1 * cls.enter2 * cls.enter3 * cls.answer
        item_2 = cls.hcf_ab * cls.hcf_bc * cls.hcf_ca
        print("LCM is>", item//item_2)

    
```

```python
    @classmethod
    def final(cls):
        #if cls.frc == 'y':
         #   cls.entry_frc_1()
          #  cls.frc_1_hcf()
        if cls.yn == 2:
            Hcf_Lcm.entry()
            Hcf_Lcm.hcf_2()
            Hcf_Lcm.lcm_2()
        elif cls.yn == 3:
            Hcf_Lcm.entry_2()
            Hcf_Lcm.hcf_3()
            Hcf_Lcm.lcm_3()

 
```
